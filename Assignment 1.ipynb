# Fractions

```python
class fraction:
    def __init__(self , s , m):
        self.s = s
        self.m = m

    def sum(self , f) :
        if self.m == f.m : #اگر کسر ها هم مخرج باشند
            result_s = self.s + f.s
            result_m = self.m
            result = fraction(result_s, result_m)
            return result
        else: #اگر کسر ها هم مخرج نباشند
            result_s = self.s * f.m + f.s * self.m
            result_m = self.m * f.m
            result = fraction(result_s, result_m)
            return result

    def mul(self , f) :
        result_s = self.s * f.s
        result_m = self.m * f.m
        result = fraction(result_s, result_m)
        return result

    def sub(self , f) :
        if self.m == f.m : #اگر کسر ها هم مخرج باشند
            result_s = self.s - f.s
            result_m = self.m
            result = fraction(result_s, result_m)
            return result
        else: #اگر کسر ها هم مخرج نباشند
            result_s = self.s * f.m - f.s * self.m
            result_m = self.m * f.m
            result = fraction(result_s, result_m)
            return result

    def div(self , f) :
        result_s = self.s * f.m
        result_m = self.m * f.s
        result = fraction(result_s, result_m)
        return result

    def show(self) :
        print(self.s, "/", self.m)
        
    def fraction_to_number(self) : # تبدیل کسر به عدد
        if self.m == 0 :
            return "Error"
        else :
            print(self.s / self.m)
        """else:
            result = self.s / self.m
            return result"""
    

a = int(input("surate kasre avval ra vared konid:"))
b = int(input("makhraje kasre avval ra vared konid:"))
f1 = fraction(a,b)
f1.show()

c = int(input("surate kasre dovvom ra vared konid:"))
d = int(input("makhraje kasre dovvom ra vared konid:"))
f2 = fraction(c,d)
f2.show()

# جمع دو کسر
sum_result = f1.sum(f2)
print("جمع:")
sum_result.show()
sum_result.fraction_to_number()

# تفریق دو کسر
sub_result = f1.sub(f2)
print("تفریق:")
sub_result.show()
sub_result.fraction_to_number()

# ضرب دو کسر
mul_result = f1.mul(f2)
print("ضرب:")
mul_result.show()
mul_result.fraction_to_number()

# تقسیم دو کسر
div_result = f1.div(f2)
print("تقسیم:")
div_result.show()
div_result.fraction_to_number()
```

# Time

```python
class time:
    def __init__(self, h, m, s):
        self.h=h
        self.m=m
        self.s=s

    def sum(self, t):
        h=self.h+t.h
        m=self.m+t.m
        s=self.s+t.s
        
        if s >= 60:
            s -= 60
            m += 1
            
        if m >= 60:
            m -= 60
            h += 1
        result = time(h, m, s)
        return result

    def sub(self, t):
        h = self.h - t.h
        m = self.m - t.m
        s = self.s - t.s

        if s < 0:
            s += 60
            m -= 1

        if m < 0:
            m += 60
            h -= 1

        if h < 0:
            print("Subtraction can not being calculated")
            result = time(0, 0, 0)
            return result

        result = time(h, m, s)
        return result
    
    def gmt_to_teh(self):
        s = self.s
        m = self.m + 30
        h = self.h + 3
        
        if m >= 60 :
            m -= 60
            h += 1
            
        result = time(h, m, s)
        return result
    
    def time_to_sec(self):
        total_seconds = self.h * 3600 + self.m * 60 + self.s
        return total_seconds
    
    def show(self):
        print(f"{self.h}:{self.m}:{self.s}")

class Second:
    def __init__(self, t):
        self.t = t
        self.h = 0
        self.m = 0
        self.s = 0

    def sec_to_time(self):
        if self.t > 60:
            self.s = self.t % 60
            self.m = (self.t // 60) % 60
            self.h = self.t // 3600
        else:
            self.s = self.t
        
        print(f"{self.h}:{self.m}:{self.s}")        
        
t1=time(9,15,35)
t1.show()

t2=time(8,40,50)
t2.show()

#جمع
s=t1.sum(t2)
a=s.show()

#تفریق
s=t1.sub(t2)
s.show()

#تبدیل زمان به ثانیه
s=t1.time_to_sec()
print("Total seconds:", s)

#تبدیل ثانیه به زمان
input_second = int(input("Enter seconds: "))
sec = Second(input_second)
sec.sec_to_time()

#تبدیل زمان گرینویچ به زمان تهران
t3=time(10,30,10)
s=t3.gmt_to_teh()
s.show()


```
